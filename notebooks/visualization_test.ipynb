```python
# test_visualization.ipynb

import numpy as np
import matplotlib.pyplot as plt
import py_utils.visualization as vz
# import py_utils.vizualization as vz

# Generate test signal
N = 128
Fs = 1000
t = np.arange(N) / Fs
signal = np.sin(2*np.pi*50*t) + 0.5*np.sin(2*np.pi*120*t)

# --- Test 1: plot_signal ---
fig, ax = vz.plot_signal(signal, title="Test Signal", xlim=[0, 50])
assert isinstance(fig, plt.Figure)
assert isinstance(ax, plt.Axes)
print("plot_signal: PASS")

# --- Test 2: plot_spectrum ---
fig, ax = vz.plot_spectrum(signal, Fs=Fs)
assert isinstance(fig, plt.Figure)
assert isinstance(ax, plt.Axes)
print("plot_spectrum: PASS")

# --- Test 3: plot_constellation ---
# Fake QPSK symbols
bits = np.random.randint(0, 2, N*2)
symbols = (2*bits[0::2]-1) + 1j*(2*bits[1::2]-1)
fig, ax = vz.plot_constellation(symbols, n_samples=200)
assert isinstance(fig, plt.Figure)
assert isinstance(ax, plt.Axes)
print("plot_constellation: PASS")

# --- Test 4: visualize wrapper ---
figs = vz.visualize(signal, plots=("time", "fft"))
assert "time" in figs and "fft" in figs
print("visualize: PASS")

print("All visualization tests completed successfully.")
```
